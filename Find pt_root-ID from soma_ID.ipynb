```python
# Import requred packages.
import pandas as pd
import numpy as np
from datetime import datetime 
```

```python
# Import a package 'fanc' to interact with EM dataset 'famale adult nerve code = fanc'. (ref. https://github.com/htem/FANC_auto_recon/wiki/)
import fanc
client = fanc.get_caveclient()
```

```python
client
```

```python
version_number = client.materialize.version
version_number
```

```python
# Available version #s
client.materialize.get_versions()
```

```python
# Confirm which version will be used.
#version_number = 1048
```

```python
time_stamp = client.materialize.get_timestamp()
time_stamp
```

```python
date=datetime.now().date()
```

```python
#Doownload a table "soma_dec2022" from FANC dataset of specified version.
soma_id_df = client.materialize.query_table('somas_dec2022', materialization_version=version_number)
soma_id_df
```

```python
#Doownload a table "soma_dec2022" from FANC dataset of specified version.
# Choose only requred colums (here, id(soma_id), pt_root_id (segment_id in specified version).
#A table is 'DataFrame'. These id#s' type are 'int'.
#Data type of #ids will be changed from 'int' to 'str' to avoid any problem caused by merging, up/downloading  etc, later.
#Change a name of column "id" to "soma_id".

soma_id_df2 = client.materialize.query_table('somas_dec2022', select_columns = ['id', 'pt_root_id'],\
                                             materialization_version=version_number)
soma_id_df2[['id','pt_root_id']] = soma_id_df[['id','pt_root_id']].astype('str')
soma_id_df3 = soma_id_df2.rename(columns={"id": "soma_id"})
soma_id_df3
```

```python
soma_id_df3.to_csv('~/Downloads/soma_id# -> pt_root_id#_v = ' + str(version_number) +'_'+str(date)+'.csv')
```
